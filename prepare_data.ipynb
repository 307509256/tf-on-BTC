```python
raw_data = []
with open("./docs/btc_usd.csv", mode="r", encoding="utf-8") as f:
    for line in f.readlines():
        if not "å¹´" in line:
            continue
        tmp = line.split('","')
        raw_data.append({
            "time":tmp[0][1:],
            "close":float(tmp[1].replace(",","")),
            "high":float(tmp[3].replace(",","")),
            "low":float(tmp[4].replace(",",""))
        })
```
